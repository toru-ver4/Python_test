# find と findAll の違いを確認

要素が複数個あった場合、最初の1個だけ出力するのが *find()*, 全部出力するのが *findAll()*。

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
html = urlopen("https://en.wikipedia.org/wiki/Dynamic_range")
bsObj = BeautifulSoup(html, "lxml")

print(bsObj.find("table"))
print("=========")
print(bsObj.findAll("table"))

# for find_ret in bsObj.find("table"):
#     print("----------------")
#     print(find_ret)
#     print("---------------")
    
# for find_ret in bsObj.findAll("table"):
#     print("==========")
#     print(find_ret)
#     print("==============")
```
